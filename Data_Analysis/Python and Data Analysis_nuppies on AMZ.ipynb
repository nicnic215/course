```python
df = pd.read_csv("Amazon.csv")
```

```python
df["Brand Name"] = ""
```

```python
df2 = pd.read_csv("nappy_brands.csv")
```

```python
df2["Brand"] = df2["Brand"].str.upper()
BrandList = list(df2.Brand)
df["Product Name"] = df["Product Name"].str.upper()
```

```python
for x in range(0, len(df["Brand Name"])):
    for y in range(0, len(BrandList)):
        if (BrandList[y] in df["Product Name"].loc[x]) == True:
            df["Brand Name"].loc[x] = BrandList[y]
```

```python
df["Number of Reviews"].replace(',', '', regex = True, inplace = True)
df["Rating"].replace('Null', np.nan, regex = True, inplace = True)
df["Number of Reviews"].replace('Null', np.nan, regex = True, inplace = True)
df["Price"].replace('Null', np.nan, regex = True, inplace = True)
df["Price Count"].replace('Null', np.nan, regex = True, inplace = True)
df["Brand Name"].replace('', np.nan, regex = True, inplace = True)
```

```python
df["Rating"] = df["Rating"].astype(float)
df["Number of Reviews"] = df["Number of Reviews"].astype(float)
df["Price"] = df["Price"].astype(float)
df["Price Count"] = df["Price Count"].astype(float)
```
